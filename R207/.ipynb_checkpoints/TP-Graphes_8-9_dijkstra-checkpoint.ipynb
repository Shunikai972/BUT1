# TP 8-9 : implémentation de la recherche de plus coruts chemins
Dans ce TP, l'objectif est d'implémenter la recherche de plus courts chemins, avecl'algorithme de Dijkstra, dans des graphes valués. 

La représentation choisie pour les graphes est une matrice d'adjacence, qui sera représentée par un tableau de tableau (les noms des sommets sont des entiers et seront numérotés en commençant à 0). Nous considérons que le nombre de sommets du graphe ne pourra pas être modifié une fois le graphe construit.

## Création d'une fonction de conversion d'un graphe, défini par une matrice d'adjacence, vers le format graphviz

```python
def creation_graphe_value_graphviz(g):
    # création d'un graphe valué orienté
    dot = graphviz.Digraph()
    
    # ajout des sommets
    for i in range(len(g)):
        dot.node(str(i))
    
    # ajout des arcs
    for i in range(len(g)):
        for j in range(len(g)):
            if g[i][j] != math.inf:
                dot.edge(str(i), str(j), label=str(g[i][j]))
    
    return dot
```

## Fonctions fournies pour manipuler les graphes

#### Fonction qui construit un graphe vide (sans arcs), avec le nombre de sommets donné, et qui retourne la matrice d'adjacence correspondante

```python
def graphe_vide(n):
    g = []
    for i in range(n):
        l = []
        for j in range(n):
            l.append(math.inf)
        g.append(l)
    return g
```

#### Fonction qui retourne vrai si l'arc, dont les numéros de sommets extrémités sont donnés, existe

```python
def est_arc(g, i, j):
    return g[i][j] != math.inf
```


#### Fonction qui ajoute un arc au graphe, s'il n'est pas déjà présent

```python
def ajoute_arc(g, i, j, v):
    g[i][j] = v
```

#### Fonction qui ote un arc du graphe, s'il était déjà présent

```python
def ote_arc(g, i, j):
    g[i][j] = math.inf
```

#### Fonction qui retourne le nombre de sommets du graphe

```python
def nb_sommets(g):
    return len(g)
```

#### Fonction qui retourne un tableau contenant les noms des sommets successeurs du sommet donné

```python
def successeurs(g, s):
    succs = []
    for i in range(len(g)):
        if g[s][i] != math.inf:
            succs.append(i)
    return succs
```

#### Fonction qui indique si la liste des sommets donnés correspond à un chemin, dans le graphe donné

```python
def est_chemin(g, ls):
    # un chemin peut être vide ou ne contenir qu'un sommet
    if len(ls) < 2:
        return True
    
    for i in range(len(ls)-1):
        if not(est_arc(g,ls[i],ls[i+1])):
            return False
        
    return True
```

## Fonctions à implémenter pour la recherche de plus courts chemins avec l'algorithme de Dijkstra

#### `init_tab_distances(s, n)` : retourne le tableau des distances initialisé pour l'algorithme de Dijkstra, à partir du sommet `s`.

```python
def init_tab_distances(s, n):
    tab_distances = []
    for i in range(n):
        if i == s:
            tab_distances.append(0)
        else:
            tab_distances.append(math.inf)
            
    return tab_distances
```

```python
## test de la fonction init_tab_distances
tab = init_tab_distances(1, 5)
print(tab) #résultat attendu : [inf, 0, inf, inf, inf]
```

#### `init_tab_peres(n)` : initialise le tableau des pères pour l'algorithme de Dijkstra.

```python
def init_tab_peres(n):
    tab_peres = []
    
    for i in range(n):
        tab_peres.append(None)
        
    return tab_peres
```

```python
## test de la fonction init_tab_peres
tab = init_tab_peres(5)
print(tab) #résultat attendu : [None, None, None, None, None]
```

#### `sommet_dist_min(tab_distances, sommets_marques)` : retourne l'indice du sommet de `tab_distances` qui correspond à la distance minimale du tableau, en considérant uniquement les sommets non marqués.

```python
def sommet_dist_min(tab_distances, sommets_marques):
    x = -1
    min_dist = math.inf
    for i in range(len(tab_distances)):
        if i not in sommets_marques and tab_distances[i]<min_dist:
            min_dist = tab_distances[i]
            x = i
    
    return x
```

```python
## test de la fonction sommet_dist_min
tab = [1, 5, 4, 3, 6]
M = set()

s = sommet_dist_min(tab, M)
print("sommet avec dist min :", s) #résultat attendu : sommet avec dist min : 0
M.add(s)

s = sommet_dist_min(tab, M)
print("sommet avec dist min :", s) #résultat attendu : sommet avec dist min : 3
```

#### `algo_dijkstra(graphe, s)` : remplit et retourne les tableaux `tab_distances` et `tab_peres`, en utilisant l'algorithme de Dijkstra, en partant du sommet `s`.

```python
def algo_dijkstra(graphe, s):

    n = nb_sommets(graphe)
    tab_distances = init_tab_distances(s, n)
    tab_peres = init_tab_peres(n)

    sommets_marques = set()

    while len(sommets_marques) < n:
        
        x = sommet_dist_min(tab_distances, sommets_marques)
        
        if x == -1:
            break
        
        sommets_marques.add(x)

        for y in successeurs(graphe, x):
            
            if y not in sommets_marques:
                
                nouvelle_distance = tab_distances[x] + graphe[x][y]
                
                if nouvelle_distance < tab_distances[y]:
                    tab_distances[y] = nouvelle_distance
                    tab_peres[y] = x

    return (tab_distances, tab_peres)
```

```python
## test de la fonction algo_dijkstra
#print(graphe1)
print("\n ==> recherche de plus courts chemins avec l'algo de Dijkstra, à partir du sommet 0 :")
(tab_dist_0, tab_peres_0) = algo_dijkstra(graphe1, 0)
print("Tableau des distances :", tab_dist_0) #résultat attendu : Tableau des distances : [0, 6, 3, 2]
print("Tableau des pères :", tab_peres_0) #résultat attendu : Tableau des pères : [None, 2, 3, 0]
#print(graphe1)

print("\n ==> recherche de plus courts chemins avec l'algo de Dijkstra, à partir du sommet 3 :")
(tab_dist_3, tab_peres_3) = algo_dijkstra(graphe1, 3)
print("Tableau des distances :", tab_dist_3) #résultat attendu : Tableau des distances : [inf, 4, 1, 0]
print("Tableau des pères :", tab_peres_3) #résultat attendu : Tableau des pères : [None, 2, 3, None]
```

## Implémentation des fonctions bonus (cellules de code à compléter)

#### Fonction qui affiche un chemin sous la forme "0 -> 1 -> 2 -> 3", s'il existe, ou qui affiche "Pas de plus court chemin" si le chemin donné correspond à None.

```python
def affiche_chemin(chemin):
    
    if chemin is None:
        return "Pas de plus court chemin"
    
    return " -> ".join(str(x) for x in chemin)
```

```python
## test de la fonction affiche_chemin
print("chemin :", affiche_chemin([1, 2, 3, 4])) #résultat attendu : chemin : 1 -> 2 -> 3 -> 4

print("chemin :", affiche_chemin(None)) #résultat attendu : chemin : Pas de plus court chemin
```
